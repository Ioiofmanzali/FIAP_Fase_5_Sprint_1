```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Dados fornecidos
dados = {
    "Hierarquia": ["MU"] * 21,
    "Nome": ["Lucas do Rio Verde (MT)"] * 21,
    "Area plantada (ha)": [140000, 100290, 146248, 147057, 175073, 160480, 138000, 165800, 135330, 165263, 180440, 142000, 160450, 180450, 183075, 172600, 165600, 150000, 170000, 163000, 165000],
    "Area colhida(ha)": [140000, 100290, 146248, 147057, 175073, 160480, 138000, 165800, 135330, 165263, 180440, 142000, 160450, 177450, 183075, 172600, 165600, 150000, 170000, 163000, 165000],
    "Quantidade produzida": [588000, 332030, 529326, 596030, 709221, 691360, 809370, 794016, 601839, 1089710, 1207980, 765600, 866025, 851220, 1262010, 1099380, 1041660, 1008000, 897600, 1058100, 1175625],
    "Rendimento medio (Kg/ha)": [4200, 3310, 3619, 4053, 4051, 4308, 5865, 4788, 4447, 6594, 6695, 5392, 5397, 4797, 6893, 6370, 6290, 6720, 5280, 6491, 7125],
    "Valor da producao(R$X1000)": [107604, 72383, 97396, 101325, 200710, 221235, 186155, 111162, 225690, 301577, 260259, 143948, 195332, 425304, 246622, 365687, 384373, 591293, 934402, 1237021, 655999],
    "Ano": [2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]
}

# Criar DataFrame
df = pd.DataFrame(dados)

# Converter colunas numéricas para float (caso estejam como strings)
colunas_numericas = ["Area plantada (ha)", "Area colhida(ha)", "Quantidade produzida", "Rendimento medio (Kg/ha)", "Valor da producao(R$X1000)"]
for coluna in colunas_numericas:
    df[coluna] = df[coluna].astype(float)

# 1. Estatísticas Descritivas
estatisticas = df[colunas_numericas].describe()
print("Estatísticas Descritivas:")
display(estatisticas)

# 2. Tendências Temporais (crescimento/declínio ao longo dos anos)
df['Ano'] = df['Ano'].astype(int)
df.set_index('Ano', inplace=True)

# Calcular variação percentual anual
for coluna in colunas_numericas:
    df[f'{coluna}_Variacao_Percentual'] = df[coluna].pct_change() * 100

print("\nVariação Percentual Anual:")
display(df[[f'{coluna}_Variacao_Percentual' for coluna in colunas_numericas]])

# 3. Soma Acumulada
for coluna in colunas_numericas:
    df[f'{coluna}_Soma_Acumulada'] = df[coluna].cumsum()

print("\nSoma Acumulada:")
display(df[[f'{coluna}_Soma_Acumulada' for coluna in colunas_numericas]])

# 5. Correlações entre as variáveis
correlacoes = df[colunas_numericas].corr()
print("\nMatriz de Correlações:")
display(correlacoes)

# 6. Análise Gráfica
plt.figure(figsize=(15, 10))

# Gráfico de linhas para as colunas numéricas ao longo dos anos
for coluna in colunas_numericas:
    plt.plot(df.index, df[coluna], label=coluna)
plt.title('Evolução das Variáveis ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('Valor')
plt.legend()
plt.grid()
plt.show()

# Gráfico de correlações
plt.figure(figsize=(8, 6))
sns.heatmap(correlacoes, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlações')
plt.show()
```
